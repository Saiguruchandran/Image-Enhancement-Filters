```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('cat_image.jpeg')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.title('Original Image')
plt.show()
```

```python
average_filter = cv2.blur(image_rgb, (10, 10))
```

```python
plt.imshow(average_filter)
plt.title('Average Filter')
plt.show()
```

```python
kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]]) / 16
```

```python
weighted_filter = cv2.filter2D(image_rgb, -1, kernel)
```

```python
plt.imshow(weighted_filter)
plt.title('Weighted Filter')
plt.show()
```

```python
gaussian_filter = cv2.GaussianBlur(image_rgb, (29, 29), 0)
```

```python
plt.imshow(gaussian_filter)
plt.title('Gaussian Filter')
plt.show()
```

```python
median_filter = cv2.medianBlur(image_rgb, 15)
```

```python
plt.imshow(median_filter)
plt.title('Median Filter')
plt.show()
```

```python
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
```

```python
laplacian_sharpen = cv2.filter2D(image_rgb, -1, sharpen_kernel)
```

```python
plt.imshow(laplacian_sharpen)
plt.title('Laplacian Filter')
plt.show()
```

```python
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
laplacian_operator = cv2.Laplacian(gray, cv2.CV_64F)
laplacian_operator = np.uint8(np.absolute(laplacian_operator))
```

```python
plt.figure(figsize = (18, 6))
plt.subplot(131); plt.imshow(image_rgb); plt.title('Input Image')
plt.subplot(132); plt.imshow(gray, cmap='gray'); plt.title('Gray_image')
plt.subplot(133); plt.imshow(laplacian_operator,cmap='gray'); plt.title('Output Image ( Laplacian_filter)')
plt.show()
```

```python
plt.figure(figsize=(15, 8))

plt.subplot(2, 4, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')


plt.subplot(2, 4, 2)
plt.imshow(average_filter)
plt.title("Averaging Filter")
plt.axis('off')

plt.subplot(2, 4, 3)
plt.imshow(weighted_filter)
plt.title("Weighted Average")
plt.axis('off')

# Gaussian Filter
plt.subplot(2, 4, 4)
plt.imshow(gaussian_filter)
plt.title("Gaussian Filter")
plt.axis('off')

plt.subplot(2, 4, 5)
plt.imshow(median_filter)
plt.title("Median Filter")
plt.axis('off')


plt.subplot(2, 4, 6)
plt.imshow(laplacian_sharpen)
plt.title("Laplacian Sharpen")
plt.axis('off')

# Grayscale Image
plt.subplot(2, 4, 7)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')


plt.subplot(2, 4, 8)
plt.imshow(laplacian_operator, cmap='gray')
plt.title("Laplacian Operator")
plt.axis('off')

plt.show()
```
